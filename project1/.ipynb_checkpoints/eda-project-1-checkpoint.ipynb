## 1. Импорт данных и работа с данными

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
````

````python
currency = pd.read_csv("foreign_exchange_rates.csv")
currency.head(5)
````

````python
currency.tail(5)
````

````python
from datetime import datetime
def quarter(x):
    if x=='January' or x=='February' or x=='March':
        return 1
    elif x=='April' or x=='May' or x=='June':
        return 2
    elif x=='July' or x=='August' or x=='September':
        return 3
    elif x=='October' or x=='November' or x=='December':
        return 4
currency['Time Serie'] = currency['Time Serie'].apply(lambda x: pd.to_datetime(x))
currency['Day'] = currency['Time Serie'].apply(lambda x: x.day)
currency['Week Day'] = currency['Time Serie'].apply(lambda x: datetime.strftime(x, '%A'))
currency['Month'] = currency['Time Serie'].apply(lambda x: datetime.strftime(x, '%B'))
currency['Quarter'] = currency['Month'].apply(lambda x: quarter(x))
currency['Year'] = currency['Time Serie'].apply(lambda x: x.year)

cols = ['AUSTRALIA - AUSTRALIAN DOLLAR/US$','EURO AREA - EURO/US$', 'NEW ZEALAND - NEW ZELAND DOLLAR/US$',
       'UNITED KINGDOM - UNITED KINGDOM POUND/US$', 'BRAZIL - REAL/US$',
       'CANADA - CANADIAN DOLLAR/US$', 'CHINA - YUAN/US$',
       'HONG KONG - HONG KONG DOLLAR/US$', 'INDIA - INDIAN RUPEE/US$',
       'KOREA - WON/US$', 'MEXICO - MEXICAN PESO/US$',
       'SOUTH AFRICA - RAND/US$', 'SINGAPORE - SINGAPORE DOLLAR/US$',
       'DENMARK - DANISH KRONE/US$', 'JAPAN - YEN/US$',
       'MALAYSIA - RINGGIT/US$', 'NORWAY - NORWEGIAN KRONE/US$',
       'SWEDEN - KRONA/US$', 'SRI LANKA - SRI LANKAN RUPEE/US$',
       'SWITZERLAND - FRANC/US$', 'TAIWAN - NEW TAIWAN DOLLAR/US$',
       'THAILAND - BAHT/US$']

currency.drop(columns=['Unnamed: 0'], inplace=True)
for col in cols:
    for i in range(len(currency[col])):
        if i != 0 and currency[col][i] == 'ND':
            currency.loc[i, col] = currency.loc[i-1, col]
        elif i == 0 and currency[col][i] == 'ND':
            currency.loc[i, col] = currency.loc[i+1, col]
for col in cols:
    currency[col] = currency[col].apply(lambda x: float('%.2f'%float(x)))
currency.tail(5)
````

## 2. Визуализация и анализ

````python
plt.rcParams['figure.figsize'] = (18,15)
sns.heatmap(currency[cols].corr(), cmap='coolwarm_r', annot=True);
````

````python
plt.rcParams['figure.figsize'] = (20,5)
sns.set(style='whitegrid')
sns.set_context('notebook')
h = sns.FacetGrid(data=currency, col='Month', col_wrap=6)
h.map_dataframe(sns.lineplot, x='EURO AREA - EURO/US$', y='DENMARK - DANISH KRONE/US$', color='red')
plt.tight_layout()
plt.show()
sns.histplot(data=currency['EURO AREA - EURO/US$'], kde=True, color='blue')
plt.show()
sns.histplot(data=currency['DENMARK - DANISH KRONE/US$'], kde=True, color='blue')
plt.show()
````

````python
sns.set(style='whitegrid')
sns.set_context('notebook')
fig = plt.figure(figsize=(30,25))
axs = []
for i in range(1, len(cols)+1):
    axs.append(fig.add_subplot(int(len(cols)/4+1), 4, i))
for j in range(len(cols)):
    sns.lineplot(data=currency[(currency['Time Serie']>='2008-01-01')&
                              (currency['Time Serie']<='2013-01-01')], 
                 x='Time Serie', y=cols[j], ax=axs[j], color='olive')
plt.tight_layout()
````

````python
plt.rcParams['figure.figsize'] = (10,4)
sns.set(style='darkgrid')
sns.set_context('paper')
for year in currency[(currency['Year']>=2008)&(currency['Year']<=2013)]['Year'].unique():
    h1 = sns.FacetGrid(data=currency[currency['Year']==year], col='Quarter', col_wrap=4)
    h1.map_dataframe(sns.violinplot, y='CHINA - YUAN/US$', palette='Purples')
    h1.set_axis_labels(f'Year - {year}')
    plt.show()
````
