````python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
````

````python
data=pd.read_csv('houses_to_rent.csv')
````

````python
data.drop(columns=['Unnamed: 0'], inplace=True)
````

````python
prices = ['hoa', 'rent amount', 'property tax', 'fire insurance', 'total']
````

````python
for i in prices:
    data[i] = data[i].apply(lambda x: int(x.replace('R$',
                                                '').replace(',','').replace('Sem info','0').replace('Incluso','0')))

data['city'] = data['city'].apply(lambda x: 'YES' if x==1 else 'NO')
````

````python
data
````

````python
sns.set(style='whitegrid')
sns.set_context('notebook')
sns.lineplot(data=data, x='rooms', y='area');
````

````python
sns.set(style='whitegrid')
sns.set_context('paper')
h = sns.FacetGrid(data=data, col='rooms', col_wrap=3)
h.map_dataframe(sns.barplot, y='area', hue='animal', color='green')
plt.tight_layout();
````

````python
sns.set(style='darkgrid')
sns.set_context('notebook')
sns.boxplot(data=data, y='rooms', color='yellow');
````

````python
data['floor'] = data['floor'].apply(lambda x: int(x.replace('-', '0')))
plt.rcParams['figure.figsize'] = (10,4)
sns.set(style='whitegrid')
sns.set_context('paper')
sns.scatterplot(data=data, x='floor', y='rooms', hue='animal')
plt.xticks(range(min(data['floor']),max(data['floor']), 2))
sns.despine()
````

````python
plt.rcParams['figure.figsize'] = (10,4)
sns.set(style='whitegrid')
sns.set_context('paper')
sns.countplot(data=data, x='rooms', hue='animal')
sns.despine()
````

````python
plt.rcParams['figure.figsize'] = (10,4)
sns.set(style='darkgrid')
sns.set_context('paper')
sns.countplot(data=data, x='rooms', hue='furniture', palette=['blue', 'red'])
sns.despine()
````

````python
plt.rcParams['figure.figsize'] = (10,4)
sns.set(style='white')
sns.set_context('notebook')
sns.barplot(data=data, x='rooms', y='total');
````

````python
plt.rcParams['figure.figsize'] = (7,7)
sns.heatmap(data[prices].corr()[['rent amount']].sort_values(by='rent amount',
                                                             ascending=False), annot = True, cmap='coolwarm');
````

````python
sns.set(style='darkgrid')
h1 = sns.FacetGrid(data=data, col='rooms', col_wrap=3)
h1.map_dataframe(sns.barplot, y='parking spaces', hue='bathroom', color='yellow')
sns.despine()
````

````python
plt.rcParams['figure.figsize'] = (10,7)
sns.heatmap(data[['area','rooms', 'bathroom', 'parking spaces', 'floor', 'total']].corr(), cmap='OrRd_r');
````
